## Data Cleaning
In this file we have cleaned and checked if we need to perform any data preprocessing in the scraped data.

```python
#importing required modules
import pandas as pd
import numpy as np
```

```python
#reading 
data = pd.read_csv('data/ODI_Players_runs.csv',sep=',')
```

```python
#creating a copy of dataset
data_temp = data.copy()
```

```python
data_temp.head()
```

```python
data_temp.shape
```

```python
new = data_temp["Name"].str.split("(", n = 1, expand = True)
```

```python
name = new[0]
```

```python
country = new[1].str.split(")", n = 1, expand = True)[0]
```

```python
data_temp.insert(loc=0, column='Name_of_player', value=name)
data_temp.insert(loc=1, column='Country', value=country)
```

```python
data_temp.drop(['Name'],axis=1,inplace=True)
```

```python
data_temp.head()
```

```python
data_temp.drop_duplicates(inplace=True)
```

```python
data_temp.to_csv('data/ODI_Players_runs_Cleaned.csv',sep=',',index=False)
```

```python
odi_players = data_temp[['Name_of_player','Country']]
```

```python
odi_players.to_csv('data/ODI_Players.csv',sep=',',index=False)
```
